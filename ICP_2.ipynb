```python
print("Roll Number:24EU02026")
n=int(input("Enter number of elements:"))
arr=list(map(int,input("Enter elements:").split()))
d=int(input("Enter difference:"))
found=False

for i in range(n-1):
    for j in range(i+1,n):
        if abs(arr[i]-arr[j])==d:
            print(arr[i],arr[j])
            found=True
if not found:
    print(-1)

```

```python
print("Roll Number:24EU02026")
n=int(input("Enternumberofbuildings:"))
h=list(map(int,input("Enterheights:").split()))
water=0
for i in range(1,n-1):
    left=max(h[:i+1])
    right=max(h[i:])
    water=water+min(left,right)-h[i]
print("TotalRainwaterTrapped=",water,"units")

```

```python
print("RollNumber:24EU02026")
n=int(input("Enternumberofelements:"))
arr=list(map(int,input("Enterelements:").split()))
maxsum=arr[0]
for i in range(n):
    s=0

    for j in range(n):

        s=s+arr[(i+j)%n]
        if s>maxsum:
            maxsum=s
print("MaximumCircularSubarraySum=",maxsum)
```

```python
print("RollNumber:24EU02026")
n=int(input("Enternumberofdays:"))
price=list(map(int,input("Enterprices:").split()))
minprice=price[0]
profit=0
for i in range(1,n):
    if price[i]-minprice>profit:
        profit=price[i]-minprice
    if price[i]<minprice:
        minprice=price[i]
print("MaximumProfit=",profit)

```

```python
print("RollNumber:24EU02026")
n=int(input("Enternumberofchildren:"))
r=list(map(int,input("Enterratings:").split()))
c=[1]*n
for i in range(1,n):
    if r[i]>r[i-1]:
        c[i]=c[i-1]+1
for i in range(n-2,-1,-1):
    if r[i]>r[i+1] and c[i]<=c[i+1]:
        c[i]=c[i+1]+1
print("Minimum Candies Required=", sum (c))
```

```python
print("RollNumber:24EU02026")
s1=input("Enterfirststring:")
s2=input("Entersecondstring:")
if len(s1)!=len(s2):
    print("NotIsomorphic")
else:
    d1={}
    d2={}
    flag=True
    for i in range(len(s1)):
        if s1[i] in d1:
            if d1[s1[i]]!=s2[i]:
                flag=False
                break
        else:
            d1[s1[i]]=s2[i]
        if s2[i] in d2:
            if d2[s2[i]]!=s1[i]:
                flag=False
                break
        else:
            d2[s2[i]]=s1[i]
    if flag:
        print("Isomorphic")
    else:
        print("NotIsomorphic")
```
