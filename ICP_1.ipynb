```python
print("Roll Number: 24EU02026")
n = int(input("Enter a number: "))
c = 0
for i in range(1, n + 1):
    if n % i == 0:
        c += 1
if c == 2:
    print("Prime Number")
else:
    print("Not a Prime Number")

```

```python
print("Roll Number: 24EU02026")
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter elements: ").split()))
count = 0
for num in arr:
    if count == 0:

        candidate = num
    if num == candidate:
        count += 1
    else:
        count -= 1

count = 0
for num in arr:
    if num == candidate:
        count += 1

if count > n // 2:
    print("Majority Element:", candidate)
else:
    print("No Majority Element")
```

```python
print("Roll Number: 24EU02026")
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter elements: ").split()))
dup = []
for i in arr:
    if arr.count(i) > 1 and i not in dup:
        dup.append(i)
if dup:
    print("Duplicate elements:", dup)

else:
    print("No duplicate elements")

```

```python
print("Roll Number: 24EU02026")
n = int(input("Enter number of elements: "))
arr = list(map(int, input("Enter elements: ").split()))
max_count = 0
most_repeated = arr[0]
for i in arr:
    count = arr.count(i)
    if count > max_count:
        max_count = count
        most_repeated = i
print("Most repetitive number:", most_repeated)
```

```python
print("Roll Number:24EU02026")
n=int(input("Enter number of elements:"))
arr=list(map(int,input("Enter elements:").split()))
target=int(input("Enter target:"))
d={}
found=False
for i in range(n):
    x=target-arr[i]
    if x in d:
        print("Indices:",d[x],"and",i)
        print("Elements:",x,"and",arr[i])
        found=True
        break
    d[arr[i]]=i
if not found:
    print(-1)
```

```python
print("Roll Number:24EU02026")
n=int(input("Enter number of elements:"))
arr=list(map(int,input("Enter elements:").split()))
lis=[1]*n
for i in range(1,n):
    for j in range(i):
        if arr[i]>arr[j] and lis[i]<lis[j]+1:
            lis[i]=lis[j]+1
print("Length of LIS:",max(lis))
```

```python
print("Roll Number:24EU02026")
n=int(input("Enter number of elements:"))
arr=list(map(int,input("Enter elements:").split()))
left=[1]*n
right=[1]*n
for i in range(1,n):
    left[i]=left[i-1]*arr[i-1]
for i in range(n-2,-1,-1):
    right[i]=right[i+1]*arr[i+1]
for i in range(n):
    print(left[i]*right[i],end=" ")
```

```python
print("Roll Number:24EU02026")
n=int(input("Enter number of elements:"))
arr=list(map(int,input("Enter elements:").split()))
for i in range(n):
    next=-1
    for j in range(i+1,n):
        if arr[j]>arr[i]:
            next=arr[j]
            break
    print(arr[i],"->",next)
```

```python
print("Roll Number:24EU02026")
n=int(input("Enter number of elements:"))
arr=list(map(int,input("Enter elements:").split()))
target=int(input("Enter target:"))
found=False
for i in range(n-2):
    for j in range(i+1,n-1):
        for k in range(j+1,n):
            if arr[i]+arr[j]+arr[k]==target:
                print(arr[i],arr[j],arr[k])
                found=True
if not found:
    print(-1)
```
